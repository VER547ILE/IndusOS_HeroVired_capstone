```python
import pandas as pd

# Load each CSV file into a DataFrame
df1 = pd.read_csv('AppleStore1.csv')
df2 = pd.read_csv('AppleStore2.csv')
df3 = pd.read_csv('AppleStore3.csv')
df4 = pd.read_csv('AppleStore4.csv')

# Concatenate the DataFrames into one
concatenated_df = pd.concat([df1, df2, df3, df4], ignore_index=True)
```

```python
concatenated_df
```

```python
import pandas as pd
import re

# Load your concatenated DataFrame
# Assuming you already have 'concatenated_df'

# Define a function to remove special characters
def clean_app_name(app_name):
    # Use regular expression to remove special characters
    cleaned_name = re.sub(r'[^\w\s]', '', app_name)
    return cleaned_name

# Apply the clean_app_name function to the 'App Name' column
concatenated_df['App Name'] = concatenated_df['App Name'].apply(clean_app_name)

# Now, 'concatenated_df' contains app names with special characters removed
```

```python
concatenated_df
```

```python
# Assuming 'Version' is the column containing version numbers
concatenated_df['Version'] = concatenated_df['Version'].astype(str)
```

```python
concatenated_df
```

```python
# Save the combined DataFrame as an XLSX file
concatenated_df.to_excel('combined_dataset2.xlsx', index=False)

# The 'combined_dataset1.xlsx' file will be saved in the current directory.
```

```python
import pandas as pd

# Load the data from the saved Excel file
combined_df = pd.read_excel('combined_dataset1.xlsx')

# Find and isolate duplicate entries
duplicates_df = combined_df[combined_df.duplicated(subset='App Name', keep=False)]

# 'duplicates_df' now contains all rows with duplicate 'App Name' entries

# Display or export 'duplicates_df' to review the duplicates
print(duplicates_df)
```
